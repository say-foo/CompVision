```python
import cv2
import numpy as np
```

```python
#MENAMPILKAN OBJEK GARIS

background = np.zeros((400,400,3)).astype(np.uint8)

#horizontal line (red)
cv2.line(background,
         (100,350),
         (300,350),
         (50,0,255),
         3,
         cv2.FILLED)

#vertical line (green)
cv2.line(background,
         (50,100),
         (50,300),
         (25,255,0),
         9,
         cv2.LINE_8)

#garis miring (pink)
cv2.line(background,
         (250,350),
         (250,100),
         (255,0,255),
         5,
         cv2.LINE_4)

#garis miring (tosca)
cv2.line(background,
         (300,300),
         (280,100),
         (100,127,0),
         5,
         cv2.LINE_AA)

cv2.imshow("Draw Line", background)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#MENAMPILKAN OBJEK PERSEGI & PERSEGI PANJANG

background = np.zeros((350, 350, 3)).astype(np.uint8)

#outline color
cv2.rectangle(background,
              (16,30),    
              (210,150),  
              (0,10,255), 
              5)  
#fill color
cv2.rectangle(background,
              (210,150),   
              (270,270),  
              (0,200,255),
              -1,
               cv2.LINE_AA)           

cv2.imshow("Draw Rectangle", background)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#MENAMPILKAN TEKS
background = np.zeros((300, 600, 3)).astype(np.uint8)
h, w, c = background.shape

font_types = [cv2.FONT_HERSHEY_SIMPLEX,
            cv2.FONT_HERSHEY_PLAIN,
            cv2.FONT_HERSHEY_DUPLEX,
            cv2.FONT_HERSHEY_COMPLEX,
            cv2.FONT_HERSHEY_TRIPLEX,
            cv2.FONT_HERSHEY_COMPLEX_SMALL,
            cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
            cv2.FONT_HERSHEY_SCRIPT_COMPLEX,
            cv2.FONT_ITALIC ]

texts      = ['HELLO WORLD!',
            'UNIVERSITAS CATUR INSAN CENDEKIA',
            'MOCHAMMAD SAEFFUL BAKHRI',
            'PRODI TI IS',
            'MATA KULIAH COMPUTER VISION',
            'SEMANGAT',
            'SEMANGAT',
            'SEMANGAT',
            'SEMANGAT' ]

for text, font_type in zip(texts, font_types):
    frame = background.copy()
    cv2.putText(frame, 
                text, 
                (50, 50), 	                   
                font_type,     
                0.9,                          
                (10, 255, 127),                
                1,
                cv2.LINE_AA)   
    
    cv2.imshow("Write Text", frame)
    cv2.waitKey(2000) #delay 2 second
    
cv2.destroyAllWindows()
```

```python
#MENGGAMBAR GARIS

x0, y0, xt, yt = 0, 0, 0, 0

title_window = "Draw Line"
is_draw = False
frame = np.ones((400, 400, 3)).astype(np.uint8)*255
 
def draw_line(x0, y0, xt, yt):
    background = frame.copy()
    cv2.line(background, (x0, y0), (xt, yt), (255,0,0), 15, cv2.LINE_AA)
    cv2.imshow(title_window, background)
       
def read_line(event,x,y,flags,param):
    
    global x0, y0, xt, yt, is_draw
    
    if event == cv2.EVENT_LBUTTONDOWN:
        x0, y0, xt, yt = x, y, x, y
        is_draw = True

    elif event == cv2.EVENT_MOUSEMOVE:
        xt, yt = x, y

    elif event == cv2.EVENT_LBUTTONUP:
        xt, yt = x, y
        is_draw = False
        
    if is_draw :
        draw_line(x0, y0, xt, yt)
    
    
cv2.namedWindow(title_window) 
cv2.setMouseCallback(title_window, read_line) 

draw_line(x0, y0, xt, yt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#MENGGAMBAR LINGKARAN

import math
x0, y0, r = 0, 0, 0

title_window = "Draw Circle"
is_draw = False
frame = np.ones((400, 400, 3)).astype(np.uint8)*255
 
def draw_circle(x0, y0, r):
    background = frame.copy()
    cv2.circle(background, (x0, y0), r, (255,255,200), 2, cv2.LINE_AA)
    cv2.imshow(title_window, background)
    
    
def read_circle(event,x,y,flags,param):
    
    global x0, y0, r, is_draw
    
    if event == cv2.EVENT_LBUTTONDOWN:
        x0, y0, r = x, y, 0
        is_draw = True

    elif event == cv2.EVENT_MOUSEMOVE:
        r = int(math.sqrt((x - x0)**2 + (y - y0)**2))

    elif event == cv2.EVENT_LBUTTONUP:
        r = int(math.sqrt((x - x0)**2 + (y - y0)**2))
        is_draw = False
        
    if is_draw :
        draw_circle(x0, y0, r)
    
    
cv2.namedWindow(title_window) 
cv2.setMouseCallback(title_window, read_circle) 

draw_circle(x0, y0, r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#MENGGAMBAR PERSEGI
x0, y0, xt, yt = 0, 0, 0, 0

title_window = "Draw Rectangle"
is_draw = False
frame = np.ones((400, 400, 3)).astype(np.uint8)*255
 
def draw_rectangle(x0, y0, xt, yt):
    background = frame.copy()
    cv2.rectangle(background, (x0, y0), (xt, yt), (10,100,0), -1, cv2.LINE_AA)
    cv2.imshow(title_window, background)
    
def read_rectangle(event,x,y,flags,param):
    
    global x0, y0, xt, yt, is_draw
    
    if event == cv2.EVENT_LBUTTONDOWN:
        x0, y0, xt, yt = x, y, x, y
        is_draw = True

    elif event == cv2.EVENT_MOUSEMOVE:
        xt, yt = x, y

    elif event == cv2.EVENT_LBUTTONUP:
        xt, yt = x, y
        is_draw = False
        
    if is_draw :
        draw_rectangle(x0, y0, xt, yt)
    
    
cv2.namedWindow(title_window) 
cv2.setMouseCallback(title_window, read_rectangle) 

draw_rectangle(x0, y0, xt, yt)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
